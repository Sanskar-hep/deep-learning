```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
X,Y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_clusters_per_class=1, hypercube=False, class_sep=10, n_classes=2, random_state=41)
sns.scatterplot(x=X[:,0], y=X[:,1], hue=Y)
plt.show()
```

```python
lr = 0.01

Y = np.where(Y==0,-1,1)
def perceptron(X,Y):
  # Initialize w1, w2, b locally within the function
  w1 = 1
  w2 = 1
  b = 1
# find the values of  w1, w2 and b for which the loss function is minimal
  for i in range(1000):
    for j in range(X.shape[0]):
      z = w1*X[j,0] + w2*X[j,1] + b
      if z*Y[j] <= 0:
        w1 = w1 + (lr * Y[j] * X[j,0])
        w2 = w2 + (lr * Y[j] * X[j,1])
        b = b + (lr * Y[j])

  return w1, w2, b
```

```python
w1, w2, b = perceptron(X, Y)
```

```python
# plot the points, like you can calculate the slope of Ax+By+C=0
# so m = -(A/B)  and c = -(C/B)

m = -(w1/w2)
c = -(b/w2)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c

sns.scatterplot(x=X[:,0], y=X[:,1], hue=Y)
plt.plot(x_input, y_input)
plt.ylim(-3,3)
plt.show()
```
